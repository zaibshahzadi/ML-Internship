**K-Means Clustering**

Link: https://www.geeksforgeeks.org/machine-learning/k-means-clustering-introduction/

https://www.geeksforgeeks.org/machine-learning/elbow-method-for-optimal-value-of-k-in-kmeans/

Notes:

**K‑Means Clustering**

Unsupervised learning algorithm, no labels required.

Groups data into k clusters based on similarity.

**Steps:**

Choose random centroids.

Assign each point to the nearest centroid.

Update centroid (mean of cluster points).

Repeat until stable.

**Distance measure:** usually Euclidean distance.

**Choosing k:** Elbow method is commonly used.

**Applications:** customer segmentation, image compression, anomaly detection, document clustering.

**Challenges:** selecting the right k is difficult, depends on initial centroids, sensitive to outliers.

**Hierarchical Clustering**

link: https://www.geeksforgeeks.org/machine-learning/hierarchical-clustering/

https://www.youtube.com/watch?v=zxQF8Rmpk1M

Notes:

Unsupervised learning, no labels needed.

Builds a hierarchy of clusters (tree structure).

Visualized using a dendrogram.

**Types:**

Agglomerative (bottom‑up, merge clusters).

Divisive (top‑down, split clusters).

**Linkage methods:** single, complete, average, Ward’s.

**Applications:** customer segmentation, pattern recognition, data exploration.

**Challenges:** slow for large datasets, sensitive to noise/outliers, results depend on linkage choice.

**PCA for visualization and feature reduction**

Link: https://www.youtube.com/watch?v=Dv-Kk7PDEas&t=33s

https://www.geeksforgeeks.org/data-analysis/principal-component-analysis-pca/

Notes:

Dimensionality reduction → reduces number of features.

Converts correlated variables into uncorrelated principal components.

Goal: keep maximum variance (information) with fewer dimensions.

**Steps:**

Standardize data.

Compute covariance matrix.

Find eigenvectors & eigenvalues.

Select top components.

Project data into new space.

PC1: captures maximum variance.

PC2: next best, orthogonal to PC1.

**Advantages:** removes redundancy, faster computation, better visualization.

**Challenges:** components hard to interpret, scaling required, possible info loss.

**GridSearchCV, RandomizedSearchCV**

Link: https://www.youtube.com/watch?v=w4frwjt8uCo

https://www.geeksforgeeks.org/machine-learning/comparing-randomized-search-and-grid-search-for-hyperparameter-estimation-in-scikit-learn/

Notes:

**GridSearchCV**

Exhaustive search, tries all possible combinations of hyperparameters.

Guarantees finding the best combination (within given ranges).

Slower when parameter space is large.

Good when search space is small and well‑defined.

**RandomizedSearchCV**

Random search, samples a fixed number of random combinations.

Faster than GridSearchCV for large parameter spaces.

May not test all combinations, but often finds good results quickly.

Good when search space is large or continuous.
