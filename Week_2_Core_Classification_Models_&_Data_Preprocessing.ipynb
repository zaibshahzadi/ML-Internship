**Decision Trees, k-Nearest Neighbors (kNN), Naive Bayes**
link: https://www.geeksforgeeks.org/machine-learning/top-machine-learning-algorithms-for-classification/

Notes:

**Decision Trees**
Splits data step‑by‑step like a flowchart.

Each node = a question, each leaf = final prediction.

Easy to interpret and visualize.

Works for both classification and regression.

Risk: can overfit if tree is too deep.

**k‑Nearest Neighbors (kNN)**

Predicts based on closest neighbors.

Uses distance (Euclidean) to find K nearest points.

Classification: majority vote of neighbors.

Regression: average of neighbors’ values.

Simple, no training phase needed.

Weakness: slow for large datasets, sensitive to irrelevant features.

**Naive Bayes**

Probability‑based algorithm using Bayes’ Theorem.

“Naive” because it assumes features are independent.

Calculates probability for each class, picks the highest.

Very fast, works well for text classification (spam filter, sentiment analysis).

Weakness: independence assumption is often unrealistic.


**Train/Test Split, Cross-Validation**
link: https://www.youtube.com/watch?v=v6DtYYafrWQ

https://www.geeksforgeeks.org/machine-learning/cross-validation-machine-learning/

Notes:

**Purpose:**

Tests model stability and generalization.

Prevents overfitting by validating on unseen data.

**Holdout Validation:**

Simple split (e.g., 70–30, 80–20).

Quick but less reliable.

**LOOCV (Leave-One-Out):**

Train on all data except one point.

Low bias, but high variance and very slow for large datasets.

**Stratified Cross-Validation:**

Maintains class distribution in each fold.

Useful for imbalanced datasets.

**K-Fold Cross-Validation:**

Dataset split into k folds (commonly k=10).

Train on k-1 folds, test on 1 fold, repeat k times.

More reliable than holdout.

**Repeated K-Fold:**

Repeats K-Fold multiple times with random shuffles.

Reduces randomness, improves reliability.

**Advantages:**

Better performance estimate.

Reduces overfitting.

Efficient use of data.

**Disadvantages:**

Computationally expensive.

Time-consuming (especially LOOCV).

Bias-variance tradeoff depends on number of folds.

**Feature Selection vs Dimensionality Reduction**

Link: https://medium.com/@asanjyal81/dimensionality-reduction-vs-feature-selection-e68f91aa8724

Notes:

**Feature Selection**

Pick only the important original features.

Keeps the meaning of features the same.

Methods: Chi-square, ANOVA, RFE, Lasso, Decision Trees.

Good when you want to know which features matter most.

Simple and fast.

Risk: might remove features that still have hidden value.

**Dimensionality Reduction**

Combine or transform features into fewer new ones.

Example: PCA makes new “principal components” from old features.

Focus is on capturing patterns/variance, not on meaning.

Methods: PCA, LDA, t-SNE, Autoencoders.

Good for large datasets with many correlated features (like images, text, genomics).

Risk: harder to explain results because features lose their original meaning.
