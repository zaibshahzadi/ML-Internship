Notes:

Supervised Learning

Labeled data (input + output known)

Tasks: Classification (spam detection), Regression (house price prediction)

Algorithms: Linear Regression, Logistic Regression, Decision Trees, SVM

Unsupervised Learning

Unlabeled data (patterns finding)

Tasks: Clustering (K‑Means), Dimensionality Reduction (PCA), Association Rules (Apriori)

Applications: Customer segmentation, anomaly detection

Reinforcement Learning

Agent + Environment → learns by rewards/punishments

Algorithms: Q‑Learning, SARSA, Deep Q‑Networks

Applications: Robotics, gaming AI, self‑driving cars

Semi‑Supervised Learning

Mix of labeled + unlabeled data

When labeling is costly

Applications: NLP, medical imaging, speech recognition

Self‑Supervised Learning

When model generate labels itself

Techniques: BERT (masked modeling), Contrastive Learning, Autoencoders

Applications: NLP (GPT, BERT), computer vision, speech recognition

Notes:

Problem Definition: First step is to clearly define the problem and project goals.

Data Collection: Gather relevant, high‑quality, and sufficient data.

Notes:

Bias: Error due to overly simple assumptions, leads to underfitting.

Variance: Error due to model being too sensitive to training data, leads to overfitting.

Tradeoff:

Simple models, High bias, low variance.

Complex models, Low bias, high variance.

Best model, Balance both bias and variance for good generalization.

Irreducible Error, Random noise in data that cannot be eliminated.

Managing Bias–Variance:

Cross Validation, test on different folds of data.

Regularization (L1/Lasso, L2/Ridge), control complexity.

Feature Selection, keep only relevant features.

Ensemble Methods, combine models for stability.

Adjust Model Complexity, choose a model that is neither too simple nor too complex.

Notes:

EDA Purpose: Understand dataset structure, detect patterns, anomalies, and prepare data for modeling.

Univariate Analysis: Study one variable at a time (histograms, box plots, bar charts).

Bivariate Analysis: Study relationship between two variables (scatter plots, correlation, covariance).

Multivariate Analysis: Study three or more variables together (pair plots, PCA, spatial analysis).

Handling Missing Data: Remove or impute values (mean, median, regression, KNN).

Exploring Data Characteristics: Distribution, central tendency (mean, median, mode), variability (standard deviation), skewness, outliers.

Data Transformation: Scaling (Min‑Max, Standardization), encoding categorical variables, mathematical transformations, feature engineering, aggregation.

Visualization: Use charts and plots to reveal trends, relationships, and anomalies.

Outlier Detection: Identify unusual values using IQR or Z‑score, decide whether to keep or remove.

Communicating Insights: Present findings with visuals, highlight key patterns, mention limitations, suggest next steps.

Applications: Market analysis, risk assessment, healthcare prediction, quality control, recommendation systems.
